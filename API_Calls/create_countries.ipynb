```python
import pandas as pd
```

```python
# Import Locations Data
path = "../Resources/CitiesWGeolocation.csv"
full_locations = pd.read_csv(path,index_col=0)
full_locations = full_locations.set_index('LocationID',drop=True)
full_locations
```

```python
# Separate out countries and their IDs
country_df = full_locations[['Country','CountryID']]
country_df = country_df.sort_values('CountryID')
country_df = country_df.drop_duplicates()
country_df = country_df.set_index('CountryID',drop=True)
country_df
```

```python
export_locations = full_locations[['CountryID','City','Latitude','State/Province','Longitude']]
export_locations
```

```python
# export locations
export_locations.to_csv('../Resources/location.csv')
```

```python
# Add country image attribution link
country_df['attribution'] = 'https://flagsapi.com/#quick'
country_df
```

```python
# Add country image link
country_df['image'] = "Images/" + country_df['Country']
country_df
```

```python
# Replace spaces with dashes
country_df = country_df.replace({'image':' '},{'image':'-'},regex=True)
country_df
```

```python
country_export = country_df[['attribution','Country','image']]
country_export.to_csv("../Resources/country.csv")
```
