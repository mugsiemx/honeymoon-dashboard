```python
import pandas as pd
```

```python
# import csv file
path = '../Resources/yelpdata.csv'
location_df = pd.read_csv(path,index_col=0)
location_df
```

```python
# Add activityID column for final csv
location_df['activityID'] = pd.Categorical(location_df['category']).codes
location_df
```

```python
# prepare locationActivity table for export
export_locationActivity = location_df[['activityID','LocationID']]
export_locationActivity
```

```python
# export locationActivity
export_locationActivity.to_csv('../Resources/locationActivity.csv')
```

```python
# create activityDF
activity_df = location_df[['activityID','category']]
activity_df = activity_df.drop_duplicates()
activity_df = activity_df.sort_values(by='activityID')
activity_df = activity_df.set_index('activityID',drop=True)
activity_df
```

```python
# add links and images
activity_df['link'] = "https://www.yelp.com/search?find_desc=" + activity_df['category']
activity_df['image'] = 'Images/' + activity_df['category'] + '.jpg'
activity_df
```

```python
# Locate new images
attributions_2 = pd.DataFrame([{"category":"casinos","attribution":"https://www.rawpixel.com/image/6330043/png-sticker-public-domain"},
                             {"category":"nightlife","attribution":"https://www.rawpixel.com/image/6338437/png-sticker-public-domain"},
                             {"category":"skiing","attribution":"https://www.rawpixel.com/image/6252338/png-sticker-public-domain"},
                             {"category":"sledding","attribution":"https://pixabay.com/vectors/sledge-winter-clipart-drawing-7622214/"},
                             {"category":"wineries","attribution":"https://www.rawpixel.com/image/6469258/png-sticker-public-domain"},
                             {"category":"zoos","attribution":"https://www.rawpixel.com/image/6763538/vector-sticker-public-domain-illustrations"}])
```
